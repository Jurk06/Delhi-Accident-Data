```python
ft_rate=pd.read_csv('../input/fatality-rate/fatality_rate - Sheet1.csv')
ft_rate.head()
```

```python
ft_rate["Average"]=(ft_rate["2008_ft"]+ft_rate["2009_ft"]+ft_rate["2010_ft"]+ft_rate["2011_ft"]+ft_rate["2012_ft"]+ft_rate["2013_ft"]
                    +ft_rate["2014_ft"]+ft_rate["2015_ft"]+ft_rate["2016_ft"]+ft_rate["2017_ft"])/10
```

```python
fig = px.line(ft_rate, x="DISTRICS", y="2008_ft")
fig.show()

fig = px.line(ft_rate, x="DISTRICS", y="2009_ft")

fig.show()
```
